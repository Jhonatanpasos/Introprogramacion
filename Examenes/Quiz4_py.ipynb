Punto 1 (valor 1.5)

```python
import matplotlib.pyplot as plt
Snack = []
Precio = []
MensajeSnack = "Buenas, porfavor ingrese uno de sus snacks favoritos: "
MensajePrecio = "Ahora ingrese el precio de tu snack en el mismo orden: "
for i in range (5):
  snacks= input(MensajeSnack)
  Snack.append(snacks)
print(Snack)
for i in range (5):
  precios= int(input(MensajePrecio))
  Precio.append(precios)
print(Precio)

plt.bar (Snack, Precio, width = 0.6, color = "c")
#######################
plt.title ("¿Cual es tu snack favorito?")
plt.xlabel ("Snack favorito")
plt.ylabel ("Precios aproximados")
plt.savefig ("graficoSnacksFavoritos.png")
#######################
plt.show ()
```

Punto 2 (valor 1.5): 


```python
import matplotlib.pyplot as plt 
Ciudades = []
Poblacion = []
MensajeCiudad = "Ingresa una ciudad: "
MensajePoblacion = "Ingresa el nunero de habitantes de dicha ciudad en el mismo orden: "
for i in range (5):
  ciudad = input(MensajeCiudad)
  Ciudades.append(ciudad)
print(Ciudades)
for i in range (5): 
  habitantes = input(MensajePoblacion)
  Poblacion.append(habitantes)
print(Poblacion)

maximo = Poblacion.index(max(Poblacion))
pieExplode = [0,0,0,0,0]
pieExplode [maximo] = 0.1

plt.pie (Poblacion, labels=Ciudades, explode=pieExplode, shadow=True)
plt.title ("Densidad demografica por ciudades")
plt.savefig ("GraficoCiudades.png")
plt.show()
  
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
ecgData = pd.read_csv ("ecg.csv", encoding="UTF-8", header=0, delimiter=";").to_dict()
print (ecgData.keys())
muestras = list (ecgData["muestra"].values())
voltaje = list (ecgData["valor"].values())
plt.plot(muestras, voltaje, "c")

plt.title ("Electrocardiograma: Un electrocardiograma es un equipo medico usado para registrar las señales electricas (latidos) del corazón por unidad de tiempo")
plt.xlabel("Muestra (ms)")
plt.ylabel("voltaje (mV)")
plt.savefig("ecg.png")
plt.show ()
```
