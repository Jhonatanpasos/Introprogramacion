```python
import matplotlib.pyplot as plt

tiempo = [0, 1, 2, 3, 4, 5]
sensor = [4, 5, 6, 8, 9, 10]
plt.plot(tiempo, sensor,"--,g") #si queremos cambiar la linea y el color lo hacemos dentro de las mismas comillas  y sin espacio despues de la coma#

plt.title("Grafico sensor contra el tiempo")
plt.xlabel("tiempo (s)")
plt.ylabel("voltaje (mV)")
plt.savefig("sensor.png")
plt.show()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
ecgData = pd.read_csv ("ecg.csv", encoding="UTF-8", header=0, delimiter=";").to_dict()
print (ecgData.keys())
muestras = list (ecgData["muestra"].values())
print (muestras)
voltaje = list (ecgData["valor"].values())
print (voltaje[-10:])
plt.plot(muestras, voltaje, "c")

plt.title ("Electrocardiograma")
plt.xlabel("Muestra (ms)")
plt.ylabel("voltaje (mV)")
plt.savefig("ecg.png")
plt.show ()
```
