3. Leer que es un PPG (fotopletismografia) y graficarlo

```python
import pandas as pd 
import matplotlib.pyplot as plt
ppgData = pd.read_csv ("ppg.csv", encoding="UTF-8", header=0, delimiter=";").to_dict()
print (ppgData.keys())
muestras = list (ppgData["muestra"].values())
voltaje = list (ppgData["valor"].values())
plt.plot(muestras, voltaje, "c")

plt.title ("Fotopletismografia (PPG)")
plt.xlabel("Muestra (ms)")
plt.ylabel("voltaje (mV)")
plt.savefig("ppg.png")
plt.show ()
```
