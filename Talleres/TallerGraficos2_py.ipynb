2. Desarrollar un grafico de torta que represente las cinco ciudades de Colombia. Debes sobresalir la ciudad mas grande

```python
import matplotlib.pyplot as plt 
pieLabels = ["Medellin", "Bogota", "Barranquilla", "Cartagena", "Cali"]
Habitantes = [2569000, 7181000, 1206000, 914552, 2228000]
pieExplode = [0, 0.1, 0, 0, 0]

def etiquetarElementosPorcentuales (Habitantes, labels, indicador = ": "):
  acumulador = 0
  for elemento in Habitantes:
       acumulador += elemento
  for i in range (len(labels)):
    pieLabels[i] += indicador+str(Habitantes[i]) + " de habitantes"

etiquetarElementosPorcentuales (Habitantes, pieLabels, ": ")
plt.pie (Habitantes, labels=pieLabels, explode=pieExplode, shadow=True)
plt.title ("Densidad demografica en ciudades principales de Colombia")
plt.savefig ("GraficoCiudades.png")
plt.show()
```
