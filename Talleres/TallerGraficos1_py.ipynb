1. Hacer un gráfico de barras que represente los niveles de ingresos de una persona durante el 2020 (se puede inventar los ingresos mes a mes)

```python
import matplotlib.pyplot as plt
Meses = ["Ene", "Feb", "Mar", "Abr", "May", "Jun", "Jul", "Ago", "Sep", "Oct", "Nov", "Dic"]
Ingresos = [877803, 910538, 1000000, 877803, 754850, 877803, 1250000, 1371923, 710824, 877803, 877803, 877803]
plt.bar (Meses, Ingresos, width = 0.6, color = "c")

plt.title ("Niveles de ingresos de una persona durante el 2020")
plt.xlabel ("Meses durante el año 2020")
plt.ylabel ("Nivel de ingresos alcanzados")
plt.savefig ("GraficoIngresos.png")

plt.show ()
```
